```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import matplotlib.pyplot as plt
from tensorflow.keras.layers import TextVectorization
```

## Data Setup

```python
# setup
df = pd.read_csv('forecast_data_final.csv')
df.dropna(inplace=True)
df.shape
```

```python
df.isna().sum()
```

```python
# create labels df
df_labels = df[["danger_above", "danger_treeline", "danger_below"]]
```

```python
# train_test_split both dataframes to use in all models
from sklearn.model_selection import train_test_split

# create split for all dataframes
X_train_idx, X_test_idx = train_test_split(df.index, test_size=0.2, random_state=802)

X_raw_train = df.loc[X_train_idx, 'summary_raw']
X_raw_test = df.loc[X_test_idx, 'summary_raw']

X_clean_train = df.loc[X_train_idx, 'summary_no_stop']
X_clean_test = df.loc[X_test_idx, 'summary_no_stop']

# Labels
y_train = df_labels.loc[X_train_idx]
y_test = df_labels.loc[X_test_idx]
```

## Establish Baseline to Beat

```python
from sklearn.dummy import DummyClassifier
from sklearn.metrics import accuracy_score

def evaluate_dummy(y_train_col, y_test_col, strategy='stratified'):
    dummy = DummyClassifier(strategy=strategy, random_state=12)
    dummy.fit(np.zeros_like(y_train_col), y_train_col)
    preds = dummy.predict(np.zeros_like(y_test_col))
    return accuracy_score(y_test_col, preds)

below_acc = evaluate_dummy(y_train['danger_below'], y_test['danger_below'], strategy='stratified')
treeline_acc = evaluate_dummy(y_train['danger_treeline'], y_test['danger_treeline'], strategy='stratified')
above_acc = evaluate_dummy(y_train['danger_above'], y_test['danger_above'], strategy='stratified')

print("Stratified DummyClassifier Accuracies:")
print(f"  Below:     {below_acc:.3f}")
print(f"  Treeline:  {treeline_acc:.3f}")
print(f"  Above:     {above_acc:.3f}")
print(f"  Macro Avg: {np.mean([below_acc, treeline_acc, above_acc]):.3f}")
```

## basic dense layers

```python
from sklearn.feature_extraction.text import CountVectorizer

# text pre-processing
vectorizer = CountVectorizer(binary=True, max_features=10000) 
X_clean_train_vec = vectorizer.fit_transform(X_clean_train)
X_clean_test_vec = vectorizer.transform(X_clean_test)
```

```python
# sequential model
input_dim = X_clean_train_vec.shape[1]
num_classes = 5

inputs = keras.Input(shape=(input_dim,))
x = layers.Dense(16, activation="relu")(inputs)
x = layers.Dense(16, activation='relu')(x)

output_below = layers.Dense(num_classes, activation='softmax', name='below')(x)
output_treeline = layers.Dense(num_classes, activation='softmax', name='treeline')(x)
output_above = layers.Dense(num_classes, activation='softmax', name='above')(x)

model = keras.Model(inputs=inputs, outputs=[output_below, output_treeline, output_above])
```

```python
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy', 'accuracy', 'accuracy']
)
```

```python
history = model.fit(
    X_clean_train_vec.toarray(),
    [y_train['danger_below'], y_train['danger_treeline'], y_train['danger_above']],
    validation_data=(X_clean_test_vec.toarray(), [y_test['danger_below'], y_test['danger_treeline'], y_test['danger_above']]),
    epochs=20,
    batch_size=32
)
```

```python
history_dict = history.history
history_dict.keys()
```

```python
# below treeline plots
epochs = range(1, 21)

# below treeline accuracy
acc = history_dict["below_accuracy"]
val_acc = history_dict["val_below_accuracy"]
plt.plot(epochs, acc, "r", label="Training acc")
plt.plot(epochs, val_acc, "b", label="Validation acc")
plt.title("Training and validation accuracy for below treeline")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.show()

# below treeline loss
loss_values = history_dict["below_loss"]
val_loss_values = history_dict["val_below_loss"]
plt.plot(epochs, loss_values, "r", label="Training loss")
plt.plot(epochs, val_loss_values, "b", label="Validation loss")
plt.title("Training and validation loss for below treeline")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.show()
```

```python
# treeline plots
epochs = range(1, 21)

# treeline accuracy
acc = history_dict["treeline_accuracy"]
val_acc = history_dict["val_treeline_accuracy"]
plt.plot(epochs, acc, "r", label="Training acc")
plt.plot(epochs, val_acc, "b", label="Validation acc")
plt.title("Training and validation accuracy for at treeline")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.show()

# treeline loss
loss_values = history_dict["treeline_loss"]
val_loss_values = history_dict["val_treeline_loss"]
plt.plot(epochs, loss_values, "r", label="Training loss")
plt.plot(epochs, val_loss_values, "b", label="Validation loss")
plt.title("Training and validation loss for at treeline")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.show()
```

```python
# above treeline plots
epochs = range(1, 21)

# above treeline accuracy
acc = history_dict["above_accuracy"]
val_acc = history_dict["val_above_accuracy"]
plt.plot(epochs, acc, "r", label="Training acc")
plt.plot(epochs, val_acc, "b", label="Validation acc")
plt.title("Training and validation accuracy for above treeline")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.show()

# above treeline loss
loss_values = history_dict["above_loss"]
val_loss_values = history_dict["val_above_loss"]
plt.plot(epochs, loss_values, "r", label="Training loss")
plt.plot(epochs, val_loss_values, "b", label="Validation loss")
plt.title("Training and validation loss for above treeline")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.show()
```
